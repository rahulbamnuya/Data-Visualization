```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.
import numpy as np
import pandas as pd
import plotly.express as px

# Sample data creation
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create the heatmap using Plotly Express
fig = px.imshow(
    pivot_table.values,
    labels=dict(x="Month", y="Day", color="Sales"),
    x=pivot_table.columns,
    y=pivot_table.index,
    color_continuous_scale='Viridis'
)

# Update layout for better readability
fig.update_layout(
    title='Sales Heatmap',
    xaxis_title="Month",
    yaxis_title="Day",
    height=600,
    width=800
)

# Show the plot
fig.show()

```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function
import numpy as np
import plotly.graph_objects as go

# Create grid data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))  # Example function for surface

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout for better visualization
fig.update_layout(title='3D Surface Plot', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Show the plot
fig.show()
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data
import numpy as np
from bokeh.plotting import figure, output_file, show

# Sample data
data = np.random.randn(1000)  # Generate 1000 data points with a normal distribution

# Compute histogram
hist, edges = np.histogram(data, bins=30)

# Output to a static HTML file
output_file("histogram.html")

# Create a figure
p = figure(title="Histogram Example", background_fill_color="#fafafa")

# Add a quad glyph for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Set axis labels
p.xaxis.axis_label = "Value"
p.yaxis.axis_label = "Frequency"

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource
from bokeh.transform import linear_cmap

# Prepare data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

# Output to a static HTML file
output_file("heatmap.html")

# Create a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=z_flat.min(), high=z_flat.max())

# Create the figure
p = figure(title="Heatmap Example", x_axis_label="X Axis", y_axis_label="Y Axis",
           tools="", toolbar_location=None, width=400, height=400)

# Add the heatmap rectangles using the source
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('z', 'Viridis256', z_flat.min(), z_flat.max()), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate a synthetic dataset with correlated features
np.random.seed(42)
# Create 3 correlated features
feature_1 = np.random.normal(0, 1, 100)
feature_2 = 0.8 * feature_1 + np.random.normal(0, 0.2, 100)  # Correlated with feature_1
feature_3 = 0.5 * feature_1 + 0.5 * feature_2 + np.random.normal(0, 0.1, 100)  # Correlated with both feature_1 and feature_2

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Step 2: Calculate the correlation matrix
correlation_matrix = df.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
